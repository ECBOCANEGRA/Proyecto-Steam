```python
import json
import gzip
import ast
import pandas as pd
from datetime import datetime

```

```python
#Creamos una funcion para descomprimir el archivo y desanidar la lista
def gziprd (jsonfile):
    with gzip.open(jsonfile, 'rt') as f:
        data = [json.loads(line) for line in f]
        return data
```

```python
def gziprd1 (jsonfile):
    with gzip.open(jsonfile, 'rt') as f:
        data = [ast.literal_eval(line) for line in f]
        return data
```

```python
games = gziprd ("/Users/macbookpro/Desktop/PI_Henry/PI MLOps - STEAM/steam_games.json.gz")
reviews = gziprd1 ("/Users/macbookpro/Desktop/PI_Henry/PI MLOps - STEAM/user_reviews.json.gz")
items = gziprd1 ("/Users/macbookpro/Desktop/PI_Henry/PI MLOps - STEAM/users_items.json.gz")
```
